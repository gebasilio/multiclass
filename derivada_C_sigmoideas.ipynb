# Modelo con C sigmoideas donde C es la cantidad de Clases
## Red Neuronal de 1 Capa

```python
import multiclass_helper as MH
from matplotlib import pyplot as plt
import numpy as np
```

$
J = \sum\limits_{i=1}^C{J_i} = -\sum\limits_{i=1}^C{p_i  \log(\hat{p}_i) + (1-p_i) \log(1-\hat{p}_i)}
$

donde: 
- C es la cantidad de clases
- $p_i$ solo puede valer 1 o 0. Vale 1 si pertenece a la clase $i$'esima, y 0 si no pertenece
- $\hat{p}_i$ es la estimaci√≥n de la probabilidad

```python
fig, ax = plt.subplots(1,1,figsize=(8, 8))
biases = [['$w_{0,1}^1$',  '$w_{0,2}^1$'], 
          [ '$w_{0,1}^2$', '$w_{0,2}^2$', '$w_{0,3}^2$']]
capa_1 = [[ '$w_{1,1}^1$', '$w_{1,2}^1$'],
          [ '$w_{2,1}^1$', '$w_{2,2}^1$']]
capa_2 = [['$w_{1,1}^2$', '$w_{1,2}^2$', '$w_{1,3}^2$'],
          ['$w_{2,1}^2$', '$w_{2,2}^2$', '$w_{2,3}^2$'], 
          ['$w_{3,1}^2$', '$w_{3,2}^2$', '$w_{3,3}^2$']]
weights = [np.array(capa_1),np.array(capa_2)]
MH.draw_neural_net(ax, 0.1, 0.9, 0, 0.9, [2,2,3], weights, biases)
plt.show()
```

$\large \hat{p}_i = \frac{1}{1+e^{-\hat{y}_i}} = \sigma(\hat{y}_i)$

$\hat{y}_i = w_{0,i}^{(2)}  + \sum\limits_{k=1}^K {w_{k,i}^{(2)} a_k} $  

$z_k = w_{0, k}^{(1)} + \sum\limits_{n=1}^N w_{n, k}^{(1)} X_n$  

- K es la cantidad de neuronas en la capa oculta
- N es la cantidad de features (Dimensi√≥n de entrada)

$ \large a_k = \frac{1}{1 + e^{-z_k}} = \sigma(z_k) $

$\large \frac{\delta J}{\delta w_{n,k}^{(1)}} = \sum\limits_{i=1}^C{\frac{\delta J_i}{\delta w_{n,k}^{(1)}}}$

$\large \frac{\delta J_i}{\delta w_{n,k}^{(1)}} = \frac{\delta J_i}{\delta \hat{p}_i} \frac{\delta \hat{p}_i}{\delta \hat{y}_i} \frac{\delta \hat{y}_i}{\delta a_{k}} \frac{\delta a_k}{\delta w_{n,k}^{(1)}}$

$\large \frac{\delta J_i}{\delta w_{n,k}^{(1)}} = \frac{\delta J_i}{\delta \hat{p}_i} \frac{\delta \hat{p}_i}{\delta \hat{y}_i} \frac{\delta \hat{y}_i}{\delta a_{k}} \frac{\delta a_k}{\delta z_k} \frac{\delta a_k}{\delta w_{n,k}^{(1)}}$

recordar que la derivada de la sigmoidea es:  
$\sigma(x) [1 - \sigma(x)]$

$\large \frac{\delta J}{\delta w_{n,k}^{(1)}} = \sum\limits_{i=1}^C\frac{\delta J_i}{\delta \hat{p}_i} \sigma(\hat{y}_i) [1-\sigma(\hat{y}_i)]  w_{k,i}^{(2)} \sigma(z_k)[1-\sigma(z_k)] X_n $

$\large \frac{\delta J}{\delta w_{n,k}^{(1)}} = \sum\limits_{i=1}^C\frac{\delta J_i}{\delta \hat{p}_i} \hat{p}_i [1-\hat{p}_i]  w_{k,i}^{(2)} \hat{a}_k[1-\hat{a}_k] X_n $

$ \large \frac{\delta J_i}{\delta \hat{p}_i} = \frac{1-p_i}{1-\hat{p}_i} -\frac{p_i}{\hat{p}_i} $

$\large \frac{\delta J}{\delta w_{n,k}^{(1)}} = \sum\limits_{i=1}^C\hat{p}_i [1-p_i]  w_{k,i}^{(2)} \hat{a}_k[1-\hat{a}_k] X_n -
p_i [1-\hat{p}_i]  w_{k,i}^{(2)} \hat{a}_k[1-\hat{a}_k] X_n$

$\large \frac{\delta J}{\delta w_{n,k}^{(1)}} = \sum\limits_{i=1}^C [\hat{p}_i-p_i]  w_{k,i}^{(2)} \hat{a}_k[1-\hat{a}_k] X_n$
